```python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Hotel Reservations.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
#Convert date to single column
df['arrival_date_combined'] = pd.to_datetime(
    df['arrival_year'].astype(str) + '-' +
    df['arrival_month'].astype(str) + '-' +
    df['arrival_date'].astype(str),
    errors='coerce'  # Convert invalid dates to NaT
)
```

```python
df = df.drop(columns=['arrival_year', 'arrival_month', 'arrival_date'])
```

```python
df.head()
```

```python
# Set the combined date as the index
df.set_index('arrival_date_combined', inplace=True)
```

```python
# Aggregate data to find daily booking counts
daily_bookings = df.resample('D').size()
```

```python
# analyze seasonality
decomposition = sm.tsa.seasonal_decompose(daily_bookings, model='additive', period=12)#period= 12 indicates data is monthly

# Plotting the decomposed components: observed, trend, seasonal, and residual
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 8))
decomposition.observed.plot(ax=ax1, title='Observed')
decomposition.trend.plot(ax=ax2, title='Trend')
decomposition.seasonal.plot(ax=ax3, title='Seasonal')
decomposition.resid.plot(ax=ax4, title='Residual')
plt.tight_layout()
plt.show()
```

#checking Cancellation rates

```python
total_booking = len(df)
```

```python
#count calcellations
cancelled_bookings = df[df['booking_status'] == 'Canceled'].shape[0]
```

```python
#calculate cancellation rate
cancellation_rate = (cancelled_bookings / total_booking) * 100
```

```python
print(f'Total Bookings: {total_booking}')
print(f'Cancelled Bookings: {cancelled_bookings}')
print(f'Cancellation Rate: {cancellation_rate:.2f}%')
```

```python
status_counts = df['booking_status'].value_counts()
status_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Booking Status Distribution')
plt.xlabel('Booking Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Grouping by year and month, and count booking status
monthly_yearly_status = df.groupby([df.index.year, df.index.month])['booking_status'].value_counts().unstack(fill_value=0)

# Renaming the levels of the MultiIndex
monthly_yearly_status.index.names = ['Year', 'Month']

# Format the index for better display
monthly_yearly_status.index = [f"{int(year)}-{int(month):02d}" for year, month in monthly_yearly_status.index]

print(monthly_yearly_status.index)
```

```python
# Ploting the results
monthly_yearly_status.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'])
plt.title('Booking Status by Month and Year')
plt.xlabel('Year-Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.legend(title='Booking Status')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
cancellation_rate = (monthly_yearly_status['Canceled'] / (monthly_yearly_status['Canceled'] + monthly_yearly_status['Not_Canceled'])) * 100
```

```python
plt.figure(figsize=(10, 6))
cancellation_rate.plot(kind='line', marker='o', color='orange')
plt.title('Cancellation Rates by Month and Year')
plt.xlabel('Year-Month')
plt.ylabel('Cancellation Rate (%)')
plt.xticks(rotation=45)
plt.axhline(y=50, color='red', linestyle='--', label='50% Threshold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#to see if the reason for high cancellations were prices
july_2017_data = df[(df.index.year == 2017) & (df.index.month == 7)]
```

```python
#Calculate cancellation rates per price point
price_cancellation = july_2017_data.groupby('avg_price_per_room')['booking_status'].value_counts().unstack(fill_value=0)
```

```python
price_cancellation['cancellation_rate'] = (price_cancellation['Canceled'] / (price_cancellation['Canceled'] + price_cancellation['Canceled'])) * 100
```

```python
#Visualize
plt.figure(figsize=(12, 6))
plt.scatter(price_cancellation.index, price_cancellation['cancellation_rate'], color='blue', marker='o')
plt.title('Cancellation Rate vs. Average Price Per Room in July 2017')
plt.xlabel('Average Price Per Room')
plt.ylabel('Cancellation Rate (%)')
plt.axhline(y=50, color='red', linestyle='--', label='50% Cancellation Threshold')
plt.legend()
plt.tight_layout()
plt.show()
```
